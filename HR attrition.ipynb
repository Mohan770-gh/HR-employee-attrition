```python
df['Attrition'].replace({'Yes':[1],'No':[0]},inplace=True)
df['BusinessTravel'].replace({'Travel_Frequently':[1],'Non-Travel':[0],'Travel_Rarely':[2]},inplace=True)
df['Department'].replace({'Research & Development':[1],'Human Resources':[0],'Sales':[2]},inplace=True)
```

```python
df['EducationField'].replace({'Human Resources':[0],'Life Sciences':[1],'Marketing':[3],'Medical':[4],'Technical Degree':[5],'Other':[6]},inplace=True)
df['JobRole'].replace({'Human Resources':[0],'Healthcare Representative':[1],'Laboratory Technician':[2],'Manager':[3],'Manufacturing Director':[4],
                     'Research Director':[5],'Research Scientist':[6],'Sales Executive':[7],'Sales Representative':[8]},inplace=True)
df['MaritalStatus'].replace({'Single':[2],'Married':[0],'Divorced':[1]},inplace=True)
df['OverTime'].replace({'Yes':[1],'No':[0]},inplace=True)
```

```python
df.Gender=pd.get_dummies(df.Gender,drop_first=True)
```

```python
df.head()
```

### feature selection

```python
df.drop(['EmployeeCount','EmployeeNumber',"Over18",'StandardHours'],axis=1,inplace=True)
```

```python
X=df.drop('Attrition',axis=1)
y=df['Attrition']
```

### balancing the target columun using smote

```python
from collections import Counter
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
print('unbalanced data: ',Counter(y))
X_sm,y_sm=sm.fit_resample(X,y)
print('balanced data: ',Counter(y_sm))
```

### train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,random_state=42,test_size=0.2)
```

### model_building

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
#prediction
y_pred=dt.predict(X_test)
from sklearn.metrics import classification_report,accuracy_score,f1_score
print(classification_report(y_test,y_pred))
```
